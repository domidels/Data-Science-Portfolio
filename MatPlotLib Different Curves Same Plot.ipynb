# Representing Unemployment Rate Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt
unrate = pd.read_csv('data/unrate.csv')
unrate.head(4)
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
#iloc does not include upper limit but loc does.
plt.plot(unrate['DATE'].loc[0:11], unrate['VALUE'].loc[0:11])
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

## Plotting unemployment month rate over each year from 1948 to 1952

```python
fig = plt.figure(figsize=(6, 30))
for i in range(5):
    label = str(1948 + i)
    ax = fig.add_subplot(5,1,i+1)
    ax.plot(range(0,12), unrate.loc[i*12:i*12+11, 'VALUE'])
    ax.set_title(label)
    # Position of ticks
    ax.set_xticks(range(0,12))
    # Labels of ticks
    ax.set_xticklabels(range(1, 13))
    ax.set_xlabel("Month")
    ax.set_ylabel("Unemployment Rate")
plt.show()
```

## Plotting unemployment month rate over each year from 1948 to 1952 on the same plot with different color for each year:

```python
colors = ['y', 'c', 'g', 'b', 'r']
fig = plt.figure(figsize=(6,4))
unrate['MONTH'] = unrate['DATE'].dt.month
for i in range(5):
    label = str(1948 + i)
    plt.plot(unrate.loc[0:11, 'MONTH'], unrate.loc[i*12:i*12+11, 'VALUE'], c=colors[i])
    # Last value of unemployement for the year (y axis)
    y_pos = unrate.loc[i*12+11, 'VALUE'] 
    # Adding labels after drawing the line: x_pos is fixed at 12.2 (last value on x axis is 12)
    plt.text(12.2, y_pos, label, fontsize=10, color=colors[i])
plt.xlim(1, 14)
plt.xticks(range(1, 13))
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.show()
```

## Adding a legend:

```python
colors = ['y', 'c', 'g', 'b', 'r']
fig = plt.figure(figsize=(6,4))
unrate['MONTH'] = unrate['DATE'].dt.month
for i in range(5):
    label = str(1948 + i)
    plt.plot(unrate.loc[0:11, 'MONTH'], unrate.loc[i*12:i*12+11, 'VALUE'], c=colors[i], label=label)
    # Last y value:
plt.legend(loc='upper left')
plt.show()
```
